```
a="i love india and usa"
e=a.swapcase()
q=a.upper()
w=a.lower()
r=a.find("love")
t=a.capitalize()
y=a.isupper()
u=a.islower()
i=a.isdigit()
o=a.isalpha()
p=a.isalnum()
s=a.split()
d=" ".join(i for i in a if i in "aeiou")
f=sorted(a,reverse=True)
print(e, q, w, r, t, y, u, i, o, p, s, d, f,sep='\n')
```
